```python
import pandas as pd



data = pd.read_excel("Employees.xlsx")

data.head() #displays the first 5 rows by default

data.info()

data.shape  

data.isna().sum() #shows that all the colums are clean and no cleaning needed in this case



#data.dropna(inplace=True) for dropping NA values 
#data.drop_duplicates(inplace = True) for dropping duplicates

import matplotlib.pyplot as plt

data["Gender"].value_counts().plot(kind='pie')
plt.title('pie chart for gender column')
plt.ylabel('')
plt.show

plt.hist(data["Job Rate"])
plt.title('job rates')
plt.xlabel('Rate')
plt.ylabel('count')

data["Job Rate"].describe()

data.groupby("Department")["Annual Salary"].mean().sort_values(ascending = False).head(7).plot(kind = 'bar') #To check avg salary per department for top 7
plt.title("Avg Salary per Department")
plt.xlabel('Department')
plt.ylabel('Avg Salary')

data.groupby("Center")["Monthly Salary"].mean().sort_values(ascending = False) #To check the leading region

data['Country'].unique()

data.groupby("Country")["Job Rate"].mean().sort_values(ascending = False).plot(kind = "bar")
plt.title("Job rates by country")
plt.xlabel('Countries')
plt.ylabel('Job rates')

plt.hist(data['Overtime Hours'])
plt.title("Histogram of OT")
plt.xlabel('OT hours')
plt.ylabel('Freq')

data["Overtime Hours"].describe()

data.columns


x = data[["Years","Job Rate"]]
y = data["Annual Salary"]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)  #We keep only 20% data to test, the rest is for training

len(x_train)

len(x_test)

x

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train,y_train)

predslr = lr.predict(x_test)


from sklearn.metrics import mean_absolute_error

mean_absolute_error(predslr,y_test)

import joblib
joblib.dump(lr,"linearmodel.pkl")




```
